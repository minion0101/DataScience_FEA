データフレームの加工

```python
columns_to_keep = ['食料費', '住居費', '光熱・水道費', '家具・家事用品費', '被服及び履物費', '保健医療費', '交通・通信費', '教育費', '教養娯楽費']
Cost_taeget_df = Cost_df.drop(columns=[col for col in Cost_df.columns if col not in columns_to_keep])
Cost_taeget_df = Cost_taeget_df.iloc[::-1]  # インデックスを逆順にする
Cost_taeget_df.reset_index(drop=True, inplace=True)

Cost_taeget_df
```

教育費を対象に重回帰分析

```python
#目的変数
target = '教育費'
y = Cost_taeget_df[target].astype(np.float32)

# 説明変数
explanation = list(filter(lambda x: x != target, columns_to_keep))
x = Cost_taeget_df[explanation].astype(np.float32)
x_added_constant = sm.add_constant(x)

# 重回帰分析を実施、独立変数間の分散拡大要因を確認後、分析結果を出力
model = sm.OLS(y, x_added_constant)
num_cols = model.exog.shape[1] # 説明変数の列数
vifs = [vif(model.exog, i) for i in range(0, num_cols)]
print(f'### 対象：{target} ###')
print(pd.DataFrame(vifs, index=model.exog_names, columns=['VIF']))
```

食糧費を除外

```python
columns_to_keep = ['住居費', '光熱・水道費', '家具・家事用品費', '被服及び履物費', '保健医療費', '交通・通信費', '教育費', '教養娯楽費']
Cost_taeget_df = Cost_df.drop(columns=[col for col in Cost_df.columns if col not in columns_to_keep])
Cost_taeget_df = Cost_taeget_df.iloc[::-1]  # インデックスを逆順にする
Cost_taeget_df.reset_index(drop=True, inplace=True)

Cost_taeget_df
```

重回帰分析を実行

```python
# 目的変数
target = '教育費'
y = Cost_taeget_df[target].astype(np.float32)

# 説明変数
explanation = list(filter(lambda x: x != target, columns_to_keep))
x = Cost_taeget_df[explanation].astype(np.float32)
x_added_constant = sm.add_constant(x)

# 重回帰分析を実施、独立変数間の分散拡大要因を確認後、分析結果を出力
model = sm.OLS(y, x_added_constant)
num_cols = model.exog.shape[1] # 説明変数の列数
vifs = [vif(model.exog, i) for i in range(0, num_cols)]
print(f'### 対象：{target} ###')
print(pd.DataFrame(vifs, index=model.exog_names, columns=['VIF']))
print("\n")
result = model.fit()
print(result.summary())
```
